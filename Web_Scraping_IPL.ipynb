```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
html_data=requests.get('https://www.espncricinfo.com/records/trophy/batting-most-runs-career/indian-premier-league-117').text
html_data
```

```python
Names=[]
Span=[]
Matches=[]
Innings=[]
Not_Out=[]
Runs=[]
HS=[]
Avg=[]
BF=[]
SR=[]
Cent=[]
Fifty=[]
Ducks=[]
Fours=[]
Sixes=[]
```

```python
soup = BeautifulSoup(html_data, 'html.parser')
for i in range(1,151):
  td_tags = soup.find_all('tr')[i]
  ps = []
  for td in td_tags:
    ps.append(td.text.strip())
  Names.append(ps[0])
  Span.append(ps[1])
  Matches.append(ps[2])
  Innings.append(ps[3])
  Not_Out.append(ps[4])
  Runs.append(ps[5])
  HS.append(ps[6])
  Avg.append(ps[7])
  BF.append(ps[8])
  SR.append(ps[9])
  Cent.append(ps[10])
  Fifty.append(ps[11])
  Ducks.append(ps[12])
  Fours.append(ps[13])
  Sixes.append(ps[14])
```

```python
import pandas as pd
data = {'Names': Names, 'Span': Span, 'Matches':Matches,'Innings':Innings,'Not_Out':Not_Out,'Runs':Runs,'Highest Score':HS,'Average':Avg,'Balls Faced':BF,'Strike Rate':SR,'Centuries':Cent,'Fifties':Fifty,'Ducks':Ducks,'Fours':Fours,'Sixes':Sixes}
df = pd.DataFrame(data)
df
```

```python
df['Runs'] = df['Runs'].astype(int)
df['Balls Faced'] = df['Balls Faced'].astype(int)
df['Sixes'] = df['Sixes'].astype(int)
df['Fours'] = df['Fours'].astype(int)
```

```python
df.dtypes
```

```python
df['Balls per Boundaries']=df['Balls Faced']/(df['Sixes']+df['Fours'])
```
